# 2 Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

```python
query = "SELECT *, monthname(rental_date) AS 'month', \
weekday(rental_date) AS 'weekday',\
CASE \
WHEN weekday(rental_date) <=5 then 'Workday' \
ELSE 'Weekend' \
END AS 'day_type' \
FROM rental \
LIMIT 20;"
data = pd.read_sql_query(query, engine)
data
```

# 3 Get all films with ARMAGEDDON in the title.

```python
query = 'SELECT * FROM film WHERE title LIKE %s;'
data = pd.read_sql_query(query, engine, params=('%ARMAGEDDON%',))
data
```

# 4 Get 10 the longest films.

```python
query = 'SELECT * FROM film ORDER BY length DESC LIMIT 10;'
data = pd.read_sql_query(query, engine)
data
```

# 5 How many films include Behind the Scenes content?

```python
query = 'SELECT * FROM film WHERE special_features LIKE ('%Behind the Scenes%');'
data = pd.read_sql_query(query, engine)
data
```

# 6 Which kind of movies (rating) have a mean duration of more than two hours?

```python
query = 'SELECT rating, ROUND(AVG(length)) as avg_length FROM film GROUP BY rating HAVING avg_length >= 120;'
data = pd.read_sql_query(query, engine)
data
```

# 7 Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.

```python
query = 'SELECT title, length FROM film WHERE length NOT LIKE "0" OR length = null ORDER BY length DESC;'
data = pd.read_sql_query(query, engine)
data
```
